```python
import pandas as pd
answers = pd.read_excel('/content/var123719.xlsx')
answers
```

Подготовим новую твблицу для 1 столбца

```python
plot = pd.DataFrame(answers['Choice_1'].value_counts())
plot
```

```python
plot.columns = ['Amount']
plot
```

```python
plot_sorted = plot.sort_index()
plot_sorted
```

```python
import matplotlib.pyplot as plt
plt.bar(plot_sorted.index, plot_sorted['Amount'])
plt.xticks(rotation = 'vertical')
```

Вторая часть задачи. Сводим в единное данные по всем столбцам

```python
plot_2 = pd.DataFrame(pd.DataFrame(answers.values.reshape(-1))[0].value_counts())
plot_2.columns = ['Amount']
plot_2 = plot_2.sort_index()
plot_2
```

```python
plt.pie(plot_2['Amount'], labels = plot_2.index, autopct = '%0.1f%%')
plt.show()
```
